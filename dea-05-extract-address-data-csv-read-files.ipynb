###Extract Data from Address Files
1. Demonstrate Limitations of CSV file in SELECT statement
2. Use read_files function to overcome the limitations
3. Create Address view in the Bronze layer

###1. Demonstrate Limitations of CSV file in SELECT statement

```python
%sql
SELECT * from csv.`/Volumes/gizmobox/landing/operational_data/addresses`;
```

###2. Use read_files function to overcome the limitations

```python
%sql
SELECT * from read_files('/Volumes/gizmobox/landing/operational_data/addresses',
format => 'csv',
delimiter => '\t',
header => 'true');
```

###3. Create Address view in the Bronze layer

```python
%sql
CREATE OR REPLACE VIEW gizmobox.bronze.v_addresses AS
SELECT * from read_files('/Volumes/gizmobox/landing/operational_data/addresses',
format => 'csv',
delimiter => '\t',
header => 'true');
```

```python
%sql
SELECT * FROM gizmobox.bronze.v_addresses;
```
