###2. Create external table

```python
%sql
CREATE TABLE IF NOT EXISTS gizmobox.bronze.payments
   (payment_id INTEGER, order_id INTEGER, payment_timestamp TIMESTAMP, payment_status INTEGER, payment_method STRING)
USING CSV
OPTIONS (
  heaser = "true",
  delimiter = ","
)
LOCATION 's3://databricks-s3-external-storage/landing/external_data/payments'; --That's what makes the table external
```

```python
%sql
SELECT * FROM gizmobox.bronze.payments;
```

```python
%sql
DESCRIBE EXTENDED gizmobox.bronze.payments;
```

###3. Demonstrate the effect of adding/updating/deleting files 
If files in the external locations are deleted, added, modified, the SELECT statement will fail or not consider the new data.
We need to run the following statement to refresh the external table

```python
%sql
REFRESH TALE gizmobox.bronze.payments
```

###4. Demonstrate the effect of dropping the table
When the table is dropped, the unlerlying files are still there in the external location


```python
%sql
DROP TABLE gizmobox.bronze.payments;
```
