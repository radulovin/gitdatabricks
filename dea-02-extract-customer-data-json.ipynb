###Extract data from the customer JSON file
1. Query single file
2. Query list of files using wildcard character
3. Query all the files in the folder
4. Select file metadata
5. Register files in Unity Catalog using views
6. Create temporary views
7. Create global temporary views - GLOBAL TEMPORARY VIEW is not supported on serverless compute

###1. Query single file

```python
%fs ls /Volumes/gizmobox/landing/operational_data/customers
```

```python
%sql
SELECT * FROM json.`/Volumes/gizmobox/landing/operational_data/customers/customers_2024_10.json`
```

###2. Query multiple files

```python
%sql
SELECT * FROM json.`/Volumes/gizmobox/landing/operational_data/customers/customers_2024*.json`
```

###3. Query all the files

```python
%sql
SELECT * FROM json.`/Volumes/gizmobox/landing/operational_data/customers`
```

###4. Select File Metadata

```python
%sql
SELECT _metadata.file_path as file_path, * FROM json.`/Volumes/gizmobox/landing/operational_data/customers`
```

###5. Register files in Unity Catalog using views

```python
%sql
CREATE OR REPLACE VIEW gizmobox.bronze.v_customers AS
SELECT *, _metadata.file_path as file_path FROM json.`/Volumes/gizmobox/landing/operational_data/customers`
```

```python
%sql
SELECT * FROM gizmobox.bronze.v_customers;
```

###6. Create temporary views

```python
%sql
CREATE OR REPLACE TEMPORARY VIEW tv_customers AS
SELECT *, _metadata.file_path as file_path FROM json.`/Volumes/gizmobox/landing/operational_data/customers`
```

```python
%sql
SELECT * FROM tv_customers;
```

###7. Create global temporary views

```python
%sql
CREATE OR REPLACE GLOBAL TEMPORARY VIEW tv_customers AS
SELECT *, _metadata.file_path as file_path FROM json.`/Volumes/gizmobox/landing/operational_data/customers`
```

```python
%sql
SELECT * FROM global_temp.gtv_customers;
```
