###03. Create a new catalog

```python
%sql
CREATE CATALOG IF NOT EXISTS gizmobox
   MANAGED LOCATION 's3://databricks-s3-external-storage'
   COMMENT 'This catalog is used for the gizmobox project';
```

###04. Create schemas
1. Landing
2. Bronze
3. Silver
4. Gold

```python
%sql
SELECT current_catalog();
```

```python
%sql
USE CATALOG gizmobox;

CREATE SCHEMA IF NOT EXISTS landing
    MANAGED LOCATION 's3://databricks-s3-external-storage/landing';
CREATE SCHEMA IF NOT EXISTS bronze
    MANAGED LOCATION 's3://databricks-s3-external-storage/bronze';
CREATE SCHEMA IF NOT EXISTS silver
    MANAGED LOCATION 's3://databricks-s3-external-storage/silver';
CREATE SCHEMA IF NOT EXISTS gold
    MANAGED LOCATION 's3://databricks-s3-external-storage/gold';

SHOW SCHEMAS;
```

###05. Create Volume

```python
%sql
USE CATALOG gizmobox;
USE SCHEMA landing;

CREATE EXTERNAL VOLUME IF NOT EXISTS operational_data
LOCATION 's3://databricks-s3-external-storage/landing/operational_data';
```
